`BeautifulSoup`

Download Image

- 다음 영화에서 2021년 상위 5개의 이미지만 다운로드

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://search.daum.net/search?w=tot&q=2021%EB%85%84%EC%98%81%ED%99%94%EC%88%9C%EC%9C%84&DA=MOR&rtmaxcoll=MOR"
```

- url주소로 크롤링 해보기

```python
res = requests.get(url)
res.raise_for_status()

print(res.text)
```

- 스크래핑 : BeaurifulSoup, class name = thumb_img

```python
soup = BeautifulSoup(res.text, "html.parser")
images = soup.find_all("img", attrs={"class":"thumb_img"})
print(images)
```

결과가 리스트 안에서 <>형태로 나옴

- image 주소값 추출

```python
for idx, image in enumerate(images):
    img_url = image["src"]
    print(img_url)
    
    if img_url.startswith("//"):
        img_url = "https:" + img_url
```

- img_url을 이용해서 image data 크롤링

```python
for idx, image in enumerate(images):
    img_url = image["src"]
    print(img_url)
    
    if img_url.startswith("//"):
        img_url = "https:" + img_url

    img_res = requests.get(img_url)
    img_res.raise_for_status()

    with open(r"C:\\Users\\SOP\\Desktop\\craw\\movie{}.jpg".format(idx+1), "wb") as f:
        f.write(img_res.content)
        
    if idx >= 4:
        break
```

- 5년치에 대한 사진 다운로드

```python
for year in range(2017, 2022):
    url = "https://search.daum.net/search?w=tot&q={}%EB%85%84%EC%98%81%ED%99%94%EC%88%9C%EC%9C%84&DA=MOR&rtmaxcoll=MOR".format(year)
    res = requests.get(url)
    
    soup = BeautifulSoup(res.text, "html.parser")
    images = soup.find_all("img", attrs={"class":"thumb_img"})
    
    for idx, image in enumerate(images):
        img_url = image["src"]
        print(img_url)

        if img_url.startswith("//"):
            img_url = "https:" + img_url

        img_res = requests.get(img_url)
        img_res.raise_for_status()

        with open(r"C:\\Users\\SOP\\Desktop\\craw\\movie{}_{}.jpg".format(idx+1, year), "wb") as f:
            f.write(img_res.content)

        if idx >= 4:
            break
```
