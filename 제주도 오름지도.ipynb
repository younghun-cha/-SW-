```python
m = folium.Map(
    location=[33.3684955195788, 126.52918183373025],
    zoom_start=10
)

tooltip = 'Click!!'
# 한라산
folium.Marker(
    [33.3684955195788, 126.52918183373025],
    popup = '<strong>hello world</strong>',
    tooltip=tooltip
).add_to(m)

# 제주국제공항
folium.Marker(
    [33.5104, 126.4914],
    popup='<strong>제주국제공항</strong>',
    tooltip=tooltip
).add_to(m)
# 우도
folium.Marker(
    [33.5052, 126.9542],
    popup='<strong>우도</strong>',
    tooltip=tooltip
).add_to(m)

m
```

```python
m = folium.Map(
    location=[33.3684955195788, 126.52918183373025],
    zoom_start=10
)

tooltip = 'Click!!'
# 한라산
folium.Marker(
    [33.3684955195788, 126.52918183373025],
    popup = '<strong>hello world</strong>',
    tooltip=tooltip,
    icon=folium.Icon(color='#ff0000', icon='info-sign')
).add_to(m)

# 제주국제공항
folium.Marker(
    [33.5104, 126.4914],
    popup='<strong>제주국제공항</strong>',
    tooltip=tooltip,
    icon=folium.Icon(color='green', icon='bookmark')
).add_to(m)
# 우도
folium.Marker(
    [33.5052, 126.9542],
    popup='<strong>우도</strong>',
    tooltip=tooltip
).add_to(m)

m
```
