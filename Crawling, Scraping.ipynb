```python
from urllib.request import HTTPError
from urllib.request import URLError

try:
    html = urlopen("http://www.google.com/kim.html")
except HTTPError as e:
    print(e)
else:
    print("성공")
```

```python
from urllib.request import HTTPError
from urllib.request import URLError

try:
    html = urlopen("http://www.dddsdf.com/kim.html")
except HTTPError as e:
    print(e)
except URLError as e:
    print("The server could not be found!")
else:
    print("성공")
```

**BeautifulSoup**
- parsing 하는 library

```python
import bs4

html_str = "<html><div>hello</div></html>"
bs_obj = bs4.BeautifulSoup(html_str, "html.parser")

print(type(bs_obj))
print(bs_obj)
print(bs_obj.find("div"))
```

```python
html_str = """
<html>
    <body>
        <ul>
            <li>hello</li>
            <li>bye</li>
            <li>welcome</li>
        </ul>
    </body>
</html>
"""

bs_obj = bs4.BeautifulSoup(html_str, "html.parser")
ul = bs_obj.find("ul")
li = ul.find("li")
print(ul)

print("-"*10)

print(li)

print("-"*10)

print(li.text)

print("-"*10)

lis = ul.findAll("li")
print(lis)
print(lis[1])
print(lis[1].text)
```

```python
import bs4

html_str = """
<html>
    <body>
        <ul class = "greet">
            <li>hello</li>
            <li>bye</li>
            <li>welcome</li>
        </ul>
        <ul class = "reply">
            <li>ok</li>
            <li>no</li>
            <li>sure</li>
        </ul>
    </body>
</html>
"""

bs_obj = bs4.BeautifulSoup(html_str, "html.parser")
ul = bs_obj.find("ul")
print(ul)
```

```python
html_str = """
<html>
    <body>
        <ul class = "greet">
            <li>hello</li>
            <li>bye</li>
            <li>welcome</li>
        </ul>
        <ul class = "reply">
            <li>ok</li>
            <li>no</li>
            <li>sure</li>
        </ul>
    </body>
</html>
"""

bs_obj = bs4.BeautifulSoup(html_str, "html.parser")
ul = bs_obj.find("ul", {"class":"reply"})
print(ul)
```

```python
html_str = """
<html>
    <body>
        <ul class = "ko">
            <li>
                <a href = "http://www.naver.com/">네이버</a>
            </li>
            <li>
                <a href = "http://www.daum.com/">다음</a>
            </li>
        </ul>
        <ul class = "sns">
            <li>
                <a href = "http://www.google.com/">구글</a>
            </li>
            <li>
                <a href = "http://www.facebook.com/">페이스북</a>
            </li>
        </ul>
    </body>
</html>
"""

bs_obj = bs4.BeautifulSoup(html_str, "html.parser")
atag = bs_obj.find("a")
print(atag)
print(atag['href'])
```
